```python
def update_issues_by_status_jira(jira,l,last_ticket,project):
        size = 100
        initial = 0
        while True:
            start= initial*size
            issues = jira.search_issues('project={}'.format(project),  start,size)
            if len(issues) == 0:
                break
            initial += 1

            for issue in issues:
                if(int(issue.key.split("-")[1])>int(last_ticket)):
                    for i in l.keys():
                        if i==issue.fields.status.name:
                            l[i]+=1
                else :
                    break
        return l   
```
